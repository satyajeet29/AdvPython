```python
def lexical_diversity(textInt):
    exec("text = text%d" %(textInt))
    return len(text) / len(set(text))

def percentage(count, total):
    return 100 * count / total

def percentageAdv(phrase, textInt):
    exec("text = text%d" %(textInt))
    return 100*text.count(phrase)/len(text)

def setTextVariable(textInt):
    exec("text = text%d" %(textInt))
    return text
```

```python
for i  in range(1,10):
    print(str(i)+": "+str(str(setTextVariable(i)).replace("<Text: "," ")).
                     replace(">","") +"\t hapaxes: ")
    V = set(setTextVariable(i))
    longWords = [w for w in V if len(w) >  15]
    print(sorted(longWords))
    print("\n")
```

```python
for i  in range(1,10):
    print(str(i)+": "+str(str(setTextVariable(i)).replace("<Text: "," ")).
                     replace(">","") +"\t frequently occuring: ")
    V = FreqDist(setTextVariable(i))
    frequentWords = [w for w in set(setTextVariable(i))if len(w) >  7 and V[w] > 7]
    print(sorted(frequentWords))
    print("\n")
```

### Collocations and Bigrams

```python
#Collocation:sequence of words occuring together unusually often eg: red wine
#            characteristic: resistant to substitution

#bigrams: pair of words

#Note collocations are just frequent bigrams
```

```python
for i  in range(1,10):
    print(str(i)+": "+str(str(setTextVariable(i)).replace("<Text: "," ")).
                     replace(">","") +"\t bigrams: ")
    V = (bigrams(setTextVariable(i)))
    bigramWords = [w for w in V]
    print(bigramWords)
    print("\n")
```

```python
for i  in range(1,10):
    print(str(i)+": "+str(str(setTextVariable(i)).replace("<Text: "," ")).
                     replace(">","") +"\t bigrams: ")
    V = (setTextVariable(i).collocations())
   
    try:
        collocationWords = [w for w in V]
        print(collocationWords)
    except:
        print("An exception occurred")
    print("\n")
```

##### Concordancing of text

```python
textCounter = 0
for text in nltk.corpus.gutenberg.fileids():
    textCounter+=1
    print(str(textCounter)+": "+text+" length "+str(nltk.Text(nltk.corpus.gutenberg.words(text)).concordance("surprize")))
```

```python
textCounter = 0
for fileid in nltk.corpus.gutenberg.fileids():
    textCounter+=1
    num_chars = len(gutenberg.raw(fileid))
    num_words = len(gutenberg.words(fileid))
    num_sents = len(gutenberg.sents(fileid))
    num_vocab = len(set([w.lower() for w in gutenberg.words(fileid)]))
    print(str(textCounter)+"\t"+str(int(num_chars/num_words))+"\t"+str(int(num_words/num_sents))+"\t"+str(int(num_words/num_vocab))+"\t"+str(fileid))
 # num_words/num_vocab <<--lexical diversity score                                                                                                         
```

```python
macbeth_sentences = gutenberg.sents("shakespeare-macbeth.txt")
lineCounter = 0
for sents in macbeth_sentences:
    lineCounter+=1
    print(str(lineCounter)+":::")
    for words in sents:
        print(words)
    print("\n")
```
