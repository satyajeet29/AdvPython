```python
for i  in range(1,10):
    print(str(i)+": "+str(str(setTextVariable(i)).replace("<Text: "," ")).
                     replace(">","") +"\t outcomes: "+str(FreqDist(setTextVariable(i))))
```

```python
for i  in range(1,10):
    print(str(i)+": "+str(str(setTextVariable(i)).replace("<Text: "," ")).
                     replace(">","") +"\t keys: ")
    print(FreqDist(setTextVariable(i)).keys())
    print("\n\n\n")
```

```python
for i  in range(1,10):
    print(str(i)+": "+str(str(setTextVariable(i)).replace("<Text: "," ")).
                     replace(">","") +"\t keys: ")
    print(FreqDist(setTextVariable(i)).keys()[:50])
    print("\n\n\n")
```

```python
for i  in range(1,10):
    print(str(i)+": "+str(str(setTextVariable(i)).replace("<Text: "," ")).
                     replace(">","") +"\t hapaxes: ")
    print(FreqDist(setTextVariable(i)).hapaxes())
    print("\n\n\n")
```

```python
for i  in range(1,10):
    print(str(i)+": "+str(str(setTextVariable(i)).replace("<Text: "," ")).
                     replace(">","") +"\t hapaxes: ")
    V = set(setTextVariable(i))
    longWords = [w for w in V if len(w) >  15]
    print(sorted(longWords))
    print("\n")
```

```python
for i  in range(1,10):
    print(str(i)+": "+str(str(setTextVariable(i)).replace("<Text: "," ")).
                     replace(">","") +"\t frequently occuring: ")
    V = FreqDist(setTextVariable(i))
    frequentWords = [w for w in set(setTextVariable(i))if len(w) >  7 and V[w] > 7]
    print(sorted(frequentWords))
    print("\n")
```

### Collocations and Bigrams

```python
#Collocation:sequence of words occuring together unusually often eg: red wine
#            characteristic: resistant to substitution

#bigrams: pair of words

#Note collocations are just frequent bigrams
```

```python
for i  in range(1,10):
    print(str(i)+": "+str(str(setTextVariable(i)).replace("<Text: "," ")).
                     replace(">","") +"\t bigrams: ")
    V = (bigrams(setTextVariable(i)))
    bigramWords = [w for w in V]
    print(bigramWords)
    print("\n")
```

```python
for i  in range(1,10):
    print(str(i)+": "+str(str(setTextVariable(i)).replace("<Text: "," ")).
                     replace(">","") +"\t bigrams: ")
    V = (setTextVariable(i).collocations())
   
    try:
        collocationWords = [w for w in V]
        print(collocationWords)
    except:
        print("An exception occurred")
    print("\n")
```

### Counting other things

```python
for i  in range(1,10):
    print(str(i)+": "+str(str(setTextVariable(i)).replace("<Text: "," ")).
                     replace(">","") +"\t bigrams: ")
    V = setTextVariable(i)
    fdist = FreqDist([len(w) for w in V])
    print(fdist)
    print("Number of samples:")
    print(fdist.N())
    print(fdist.keys())
    print(fdist.items())
    print(fdist.max())
    print(fdist.plot())
    print(fdist.plot(cumulative=True))
    #print(fdist.tabulate())
    print("\n")
```
