```python
import nltk
nltk.download()
import pandas as pd
import datetime as dt
```

### Extracting whatsapp chat text

```python
from nltk.corpus import PlaintextCorpusReader
corpus_root = '/Users/satyajeetpradhan/nltk_data/corpora/Whatsapp'
wordlists = PlaintextCorpusReader(corpus_root,'.*')
wordlists.fileids()
```

```python
def wordStatistics(fileidName):
    print(str(dt.datetime.now()) +" "+fileidName+" started")
    # extracting list of words
    words = [word.lower() for word in wordlists.words(fileids = fileidName)] 
    # extracting unique set of words
    word = set(words)
    # declaring a list to store data
    data = list()
    # counter to keep track of number of words
    listcounter = 0
    # nested loop through:
    for wordUnique in word:
        listcounter+=1
        wordCounter = 0
        for wordValue in words:
            if wordUnique == wordValue:
                 wordCounter+=1
        #Creating a list variable to store data
        data.append([listcounter,wordCounter,wordUnique])
    print(str(dt.datetime.now()) +" "+fileidName+" completed")
    return data
```

```python
for fileName in wordlists.fileids():
    string = fileName.split('.',2)[0]
    if fileName != '.DS_Store':
        exec(string+"= pd.DataFrame(wordStatistics(fileName),columns=['Index','Count','WordUnique'])")
```
